```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service  #automation proceess library
from time import sleep
chrome_options = Options()
# chrome_options.add_argument("--headless=new")
from selenium.webdriver.common.by import By
```

```python
import pandas as pd
```

```python
def glbl(start, end):
      
    def start_location():
        Place = driver.find_element(By.CLASS_NAME, "searchboxinput")
        Place.send_keys(start)
        submit = driver.find_element(By.XPATH, '//*[@id="searchbox-searchbutton"]')
        submit.click()

    def direction():
        sleep(5)
        direct = driver.find_element(By.XPATH, '//*[@id="QA0Szd"]/div/div/div[1]/div[2]/div/div[1]/div/div/div[4]/div[1]/button')
        direct.click()

    def end_location():
        sleep(5)
        ind = driver.find_element(By.XPATH, '//*[@id="sb_ifc50"]/input')
        ind.send_keys(end)
        search = driver.find_element(By.XPATH, '//*[@id="directions-searchbox-0"]/button[1]')
        search.click()
    
    def click_car():
        sleep(5)
        car_sym = driver.find_element(By.XPATH, '//*[@id="omnibox-directions"]/div/div[2]/div/div/div/div[2]/button')
        car_sym.click()

    def distance_finder():
        sleep(5)
        dist = driver.find_element(By.XPATH, '//*[@id="section-directions-trip-0"]/div[1]/div/div[1]/div[2]/div')
        # print(f"the distance is  {dist.text}")
        return dist.text
    
    max_attempts = 3
    for attempt in range(1, max_attempts + 1):
        try:
            driver = webdriver.Chrome(options=chrome_options)
            driver.get("https://www.google.com/maps/@31.4855012,74.3470166,15z") 
            sleep(2) 
            start_location()
            direction()
            end_location()
            click_car()
            ans = distance_finder()
            break  # If successful, exit the loop
        except Exception as e:
            print(f"Attempt {attempt} failed. Error: {str(e)}")
            if attempt < max_attempts:
                print("Retrying...")
                sleep(4)  # Adjust the sleep time if needed
            else:
                raise Exception(f"Function failed after {max_attempts} attempts.")

    driver.quit()
    return ans

def get_distance():
    df = pd.read_csv("final.csv", index_col=0)
    df['distance'] = ''
    for i in range(len(df)):
        try:
            sleep(10)
            # print(df.at[i, "0"], df.at[i, "1"], df.at[i, "distance"])
            df.at[i, "distance"] = glbl(df.at[i, "0"], df.at[i, "1"])
            print(df.at[i, "0"], "-->", df.at[i, "1"], "-->", df.at[i, "distance"])
        except:
            print(df.at[i, "0"], "-->", df.at[i, "1"], "-->", "couldn't find")
            continue
        
    

get_distance()
# start = "Lions Point, Pune"
# end = "Bushi Dam, Pune"
# glbl(start, end)
```
